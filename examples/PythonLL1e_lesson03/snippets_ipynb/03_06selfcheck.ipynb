**2. _(IPython Session)_** Show that a `SyntaxError` occurs if an `if`…`elif` statement specifies an `else` before the last `elif`.

**Answer:**

```python
grade = 80
```

```python
if grade >= 90:
    print('A')
else:
    print('Not A or B')
elif grade >= 80:
    print('B')
```

To fix that, use `elif` in all cases:

```python
if grade >= 90:
    print('A')
elif grade >= 80:
    print('B')
else:
    print('Not A or B')
```
