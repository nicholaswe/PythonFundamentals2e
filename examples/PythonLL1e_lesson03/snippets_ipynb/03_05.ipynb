# 3.5 `if` Statement

```python
grade = 85
```

```python
if grade >= 60:
    print('Passed')
```

### Suite Indentation

```python
if grade >= 60:
print('Passed')  # statement is not indented properly
```

```python
if grade >= 60:
    print('Passed')
    print('Good job!')
```

### `if` Statement Flowchart
### Every Expression Can Be Treated as `True` or `False`

```python
# In this case, 1 means TRUE, so the print happens
if 1:
    print('Nonzero values are true, so this will print')
```

```python
# In this case, 0 means FALSE, so the print doesn't happen
if 0:
    print('Zero is false, so this will not print')
```
